```python
import pygmt

fig = pygmt.Figure()
# Create the primary figure, setting the region to Indonesia, the land color
# to "lightgreen", the water to "lightblue", and adding the Indonesian
# provinces boundaries
fig.coast(
    region="ID+r3",
    land="gray",
    water="lightblue",
    shorelines="1/0.5p,black",
    borders="1/0.5p,black",
    frame="afg",
    dcw="ID+glightgreen",
)

fig.basemap(frame=["a", "+tMap of Indonesia"])

# Add a point for Jakarta
fig.plot(x=106.8, y=-6.2, style="c0.2c", color="red", pen="black")

# Add text next to the Jakarta point
fig.text(
    x=106.8 + 0.4, y=-6.2, text="Jakarta", justify="ML", font="8p,Helvetica-Bold,black"
)

with fig.inset(position="jBL+w2.5c+o0.1c", margin=0):
    # Create the inset figure, setting the region to Indonesia, the land color
    # to "gray", the water to "white", and adding the Indonesian provinces
    # boundaries
    fig.coast(
        region="g",
        projection="G120/-5/?",
        land="gray",
        water="white",
        dcw="ID+gred3",
    )
fig.show()
fig.savefig("indonesia.png", dpi=300)
```
